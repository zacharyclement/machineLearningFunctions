```python
import pandas as pd
import numpy as np
import random
import datetime
from string import ascii_lowercase
```

```python
number_of_rows = 10000
site_ids = [5,6,8,9]
```

```python
#idvisit, 1734 unique
def gen_idvisit(number_of_rows):
    '''generate a list of numbers that gets sequentially larger, but only jumps
    to the next value after a random number of entries at the current value'''
    length = number_of_rows
    current = 1
    the_list = []
    for x in range(length):
        the_list.append(current)
        random_number = random.randint(1,100)
        if random_number < 3:
            current += 1
    return the_list
```

```python
#server_time
def gen_server_time(number_of_rows):
    '''generate a list of times that increase randomly. Simulates someone randomly clicking
    around on a website'''
    now = datetime.datetime.now()
    year = now.year
    month = now.month
    day = now.day
    date = now.date
    hour = now.hour
    minute = now.minute
    second = now.second
    date_time = datetime.datetime(year, month, day, hour, minute, second)
    #print(type(start_date_time))
    
    date_list = []
    length = number_of_rows
    for x in range(length):
        #print(date_time)
        date_list.append(date_time)
        date_time = date_time + datetime.timedelta(hours=random.randint(24,48))
    
    return date_list
```

```python
def gen_idvisitor(number_of_rows):
    length_of_id = 16
    final_list = []
    for x in range(number_of_rows):
        string = ''.join(random.choice('abcdefghijklmnopqrstuvwxyz1234567890') for i in range(length_of_id))
        #print(string)
        final_list.append(string)
    return final_list
```

```python
#idaction_name, links to action table 1 - 191
def gen_idaction_name(number_of_rows):
    length = number_of_rows
    list_ = []
    for x in range(length):
        random_int = random.randint(1,191)
        #print(random_int)
        list_.append(random_int)
    return list_
```

```python
idlink_va = [x for x in range(number_of_rows)]
idsite = [random.choice(site_ids) for x in range(number_of_rows)]
idvisitor = gen_idvisitor(number_of_rows)
idvisit = gen_idvisit(number_of_rows)
server_time = gen_server_time(number_of_rows)
idaction_name = gen_idaction_name(number_of_rows)

time_spent_ref_action = []
idaction_url_ref = []
idaction_url = []
idaction_name_ref = []
```
