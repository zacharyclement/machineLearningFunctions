```python
#idvisit, 1734 unique
def gen_idvisit(number_of_rows):
    '''generate a list of numbers that gets sequentially larger, but only jumps
    to the next value after a random number of entries at the current value'''
    length = number_of_rows
    current = 1
    the_list = []
    for x in range(length):
        the_list.append(current)
        random_number = random.randint(1,100)
        if random_number < 3:
            current += 1
    return the_list
```

```python
#server_time
def gen_server_time(number_of_rows):
    '''generate a list of times that increase randomly. Simulates someone randomly clicking
    around on a website'''
    now = datetime.datetime.now()
    year = now.year
    month = now.month
    day = now.day
    date = now.date
    hour = now.hour
    minute = now.minute
    second = now.second
    date_time = datetime.datetime(year, month, day, hour, minute, second)
    #print(type(start_date_time))
    
    date_list = []
    length = number_of_rows
    for x in range(length):
        #print(date_time)
        date_list.append(date_time)
        date_time = date_time + datetime.timedelta(hours=random.randint(24,48))
    
    return date_list
```

```python
def gen_idvisitor(number_of_rows):
    length_of_id = 16
    final_list = []
    for x in range(number_of_rows):
        string = ''.join(random.choice('abcdefghijklmnopqrstuvwxyz1234567890') for i in range(length_of_id))
        #print(string)
        final_list.append(string)
    return final_list
```
