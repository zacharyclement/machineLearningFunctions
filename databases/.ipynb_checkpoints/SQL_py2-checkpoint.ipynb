```python
data = sns.load_dataset('titanic')
```

```python
data.head()
```

```python
sns.lmplot(x='age', y='fare', hue='survived', data=data, aspect=2)
```

```python
sns.lmplot(x='age', y='fare', col='survived', data=data, aspect=2)
```

```python
sns.stripplot(x='embarked', y='age', data=data, size=4, jitter=True)
```

```python
sns.swarmplot(x='survived', y='age', hue='sex', data=data, size=4)
```

```python
sns.swarmplot(x='fare', y='survived', hue='sex', data=data, size=4, orient='h')
```

```python
sns.boxplot(x='survived', y='fare', data=data)
```

```python
sns.violinplot(x='survived', y='fare', data=data)
```

```python
sns.jointplot(x='age', y='fare', data=data, kind='kde',size=6)
```
