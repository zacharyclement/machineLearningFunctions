```python
x = df_user['visit_first_action_time']
```

```python
y = df_user['visit_total_time']/60
```

```python
fig = plt.figure()
ax = fig.add_subplot(111) #row-col-num
ax.plot(x,y,ls='None', marker='o', markersize=1)
ax.set_title('Single User Engagement Trends')
ax.set_xlabel('date')
ax.set_ylabel('time spent(minutes)')

plt.axis('tight')
plt.xticks(rotation='vertical')

plt.savefig('Single UserEngagement Trends', bbox_inches='tight')
#plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

ax1.hist(df_user['visit_total_time'], bins=20)
ax1.set_title('total time')
ax1.set_xlabel('time')
ax1.set_ylabel('count')

ax2.hist(df_user['visit_total_actions'], bins=20)
ax2.set_title('total actions')
ax2.set_xlabel('actions')
ax2.set_ylabel('count')

plt.show()
```
